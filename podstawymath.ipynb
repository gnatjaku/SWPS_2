```python
#utworz macierz 10 pkt
import numpy as np
import matplotlib.pyplot as plt

# Nasze dane
points = np.array([[2, 3, 6], [3, 5, 5], [5, 7, 7], [6, 8, 8],[8,9,9],[7,10,10],[11,5,8],[12,6,7],[13,7,8],[14,8,9]])
labels = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]

#scatter and plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

#label scatter points
ax.set_title('3D Scatter Plot Example')
# Rysowanie punktÃ³w w 3D

ax.scatter(points[:,0], points[:,1], points[:,2])

for i,label in enumerate(labels):
    ax.text(points[i,0]+0.1, points[i,1]+0.1, points[i,2]+0.1, label)


plt.show()
```

```python
def calculate_distance_matrix(points, labels):
    n = len(points)
    distance_matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if i != j:
                distance_matrix[i, j] = np.sqrt((points[i, 0] - points[j, 0])**2 + (points[i, 1] - points[j, 1])**2 + (points[i, 2] - points[j, 2])**2)
    return pd.DataFrame(distance_matrix, columns=labels, index=labels)
distance_df = calculate_distance_matrix(points, labels)
print("\nðŸ“ Macierz odlegÅ‚oÅ›ci miÄ™dzy punktami:" )
print(distance_df)
```
